```python



# Step 1: Extract the categorical columns to encode
en = df[['Gender', 'Married']]

# Step 2: Apply OneHotEncoding with drop='first'
one = OneHotEncoder(drop='first', sparse_output=False)
ar = one.fit_transform(en)

# Step 3: Create column names for the encoded features
encoded_cols = one.get_feature_names_out(['Gender', 'Married'])  # e.g. ['Gender_Male', 'Married_Yes']

# Step 4: Create a new DataFrame from encoded array
encoded_df = pd.DataFrame(ar, columns=encoded_cols)

# Step 5: Drop original categorical columns from original dataset
df_new = df.drop(['Gender', 'Married'], axis=1)

# Step 6: Concatenate original data (without Gender & Married) with encoded_df
final_df = pd.concat([df_new.reset_index(drop=True), encoded_df], axis=1)

# Step 7: View final dataset
final_df
```
