```python
from sklearn.impute import SimpleImputer
```

```python

float_cols = ['LoanAmount', 'Loan_Amount_Term', 'Credit_History']
sim = SimpleImputer(strategy='mean')

# Step 3: Fit and transform only those columns
df1[float_cols] = sim.fit_transform(df1[float_cols])
df1
```
