## Normalization

scaling technique in which values are shifter and rescaled so that they end up ranging between 0 and 1  its also known as min_max scaling   
Xnew = Xi -min (X) - min(X)

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import os
import kagglehub

# Download latest version
path = kagglehub.dataset_download("ninzaami/loan-predication")

print("Path to dataset files:", path)
```

```python
file=os.listdir(path)
```

```python
file
```

```python
df=pd.read_csv(os.path.join(path, 'train_u6lujuX_CVtuZ9i (1).csv'))
```

```python
df.head(5)
```

```python
df.isnull().sum()
```

```python
df['LoanAmount'].mean()
```

```python
df['LoanAmount'].fillna(df['LoanAmount'].mean(), inplace=True)
```

```python
df.isnull().sum()
```

```python
df.describe()
```

```python
sns.distplot(df['LoanAmount'])
plt.show()
```

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
mm=MinMaxScaler()
```

```python
mm.fit_transform(df[['LoanAmount']])
```

```python
df['LoanAmount_min']=mm.fit_transform(df[['LoanAmount']])
```

```python
df.head(3)
```

```python
df.describe()
```

```python
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title('Before')

sns.distplot(df['LoanAmount'])
plt.subplot(1,2,2)

plt.title('After')

sns.distplot(df['LoanAmount_min'])
plt.show()
```
