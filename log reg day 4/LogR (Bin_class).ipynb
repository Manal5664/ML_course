```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("student_pass_data.csv")
df
```

```python
df['Pass'].replace({'no':0 , 'yes':1},inplace=True)
df
```

```python
df['Private_Tuition'].replace({'No': '0', 'Yes': '1'}, inplace=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train, y_test=train_test_split(df[['Study_Hours', 'Private_Tuition']],df['Pass'], test_size=0.2)
```
