```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv("iris.csv")
df
```

```python
df["Species"].unique()
```

```python
df['Species'].replace({'setosa':'1', 'versicolor':'2', 'virginica':'3'}, inplace=True)
```

```python
df
```

```python
df.value_counts('Species')
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test,y_train, y_test=train_test_split(df[['Sepal.Length','Sepal.Width','Petal.Length','Petal.Width']],df['Species'], test_size=0.2)
```

```python
x_train
```

```python
x_test
```

```python
len(x_train)
```

```python
len(x_test)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr= LogisticRegression()
```

```python
lr.fit(x_train, y_train)
```

```python
lr.predict(x_test)
```

```python

x_new = np.array([[6.7,3.0,5.2,2.3]])  # shape (1, 4)
prediction = lr.predict(x_new)
print(prediction)
```

```python
lr.score(x_test, y_test)
```
