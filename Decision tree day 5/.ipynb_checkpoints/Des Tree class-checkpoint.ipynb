```python
import pandas as pd
```

```python
df=pd.read_csv("play_tennis_dataset.csv")
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
Outlook=LabelEncoder()
Temperature=LabelEncoder()
Humidity=LabelEncoder()
Windy=LabelEncoder()
Play=LabelEncoder()
```

```python
df['Outlook']=Outlook.fit_transform(df['Outlook'])
df['Temperature']=Outlook.fit_transform(df['Temperature'])
df['Humidity']=Outlook.fit_transform(df['Humidity'])
df['Windy']=Outlook.fit_transform(df['Windy'])
df['Play']=Outlook.fit_transform(df['Play'])
```

```python
df
```

```python
feature_cols=['Outlook','Temperature','Humidity','Windy']
X=df[feature_cols]
X
```

```python
y=df.Play
y
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test=train_test_split(X,y,test_size=0.2)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

```python
classifier=DecisionTreeClassifier(criterion='gini')
```

```python
classifier.fit(X_train, y_train)
```

```python
classifier.predict(X_test)
```

```python
X_test
```

```python
classifier.score(X_test,y_test)
```

```python
from sklearn import tree
tree.plot_tree(classifier)
```
