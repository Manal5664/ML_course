```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("codebreaker619/salary-data-with-age-and-experience")

print("Path to dataset files:", path)
```

```python
import os
file=os.listdir(path)
file
```

```python
df= pd.read_csv(os.path.join(path ,'Salary_Data.csv' ))
df.head()
```

```python
df['Age']=(df['Age'].astype('int64'))
```

```python
df
```

```python
df.isnull().sum()
```

```python
x=df[['YearsExperience','Age']]
y=df['Salary']
```

```python
x
```

```python
y
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y , test_size=0.2, random_state=42)
```

```python
knn=KNeighborsRegressor(n_neighbors=3)
knn.fit(x_train, y_train)
```

```python
knn.score(x_test, y_test)*100
```

```python
knn.score(x_train, y_train)*100
```

```python
for i in range(1,20):
    knn=KNeighborsRegressor(n_neighbors=i)
    knn.fit(x_train, y_train)
    print(knn.score(x_train, y_train)*100 , knn.score(x_test, y_test)*100 , i)
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(x.to_numpy(), y.to_numpy(), clf=knn)
plt.show()
```
