```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.datasets import make_moons
```

```python
x, y=make_moons(n_samples=1000 ,noise=0.2)
```

```python
dataset={'x1': x[:,0], 'x2':x[:,1],'y':y}
```

```python
df=pd.DataFrame(dataset)
```

```python
df.head(4)
```

```python
sns.scatterplot(x='x1' , y='x2', hue='y' , data=df)
plt.show()
```

```python
x_a=df.iloc[:,:-1]
```

```python
y_a=df['y']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train, x_test, y_train, y_test= train_test_split(x_a, y_a , test_size=0.2, random_state=42)
```

```python
from sklearn.ensemble import BaggingClassifier
from sklearn.ensemble import RandomForestClassifier
from sklearn.svm import SVC
```

```python
bg=BaggingClassifier(estimator=SVC(), n_estimators=50)
bg.fit(x_train, y_train)
```

```python
bg.score(x_train, y_train)*100 , bg.score(x_test, y_test)
```

```python
sv=SVC()
sv.fit(x_train, y_train)
```

```python
sv.score(x_train, y_train)*100 , sv.score(x_test, y_test)
```

```python
rf= RandomForestClassifier(n_estimators=50)
rf.fit(x_train, y_train)
```

```python
rf.score(x_test, y_test), rf.score(x_train, y_train)
```
