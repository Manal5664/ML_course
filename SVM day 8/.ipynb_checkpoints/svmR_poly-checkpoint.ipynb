```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_decision_regions
from sklearn.svm import SVR
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("dermisfit/position-salaries")

print("Path to dataset files:", path)
```

```python
import os
```

```python
file=os.listdir(path)
file
```

```python
df=pd.read_csv(os.path.join(path, 'Position_Salaries.csv'))
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(data=df, x='Level', y='Salary')
plt.show()
```

```python
x=df[['Level']]
x
```

```python
y=df['Salary']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2, random_state=42)
```

```python
svmR=SVR(kernel='rbf')
svmR.fit(x_train,y_train)
```

```python
svmR.score(x_test, y_test)*100
```

```python
svmR.score(x_train, y_train)*100
```

```python
plt.figure(figsize=(5,4))
sns.scatterplot(data=df, x='Level', y='Salary')
plt.plot(df['Level'], svmR.predict(x), color='red')
plt.show()
```
