```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
import seaborn as sns
from mlxtend.plotting import plot_decision_regions
```

```python
df=pd.read_csv('Placement.csv')
df.head(3)
```

```python
df.isnull().sum()
```

```python
sns.scatterplot(x='cgpa', y='iq', data=df , hue='placement')
plt.show()
```

```python
sns.kdeplot(data= df['cgpa'])
plt.show()
```

```python
sns.kdeplot(data= df['iq'])
plt.show()
```

```python
x=df[['cgpa', 'iq']]
y=df['placement']
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y, test_size=0.2, random_state=42)
```

```python
from sklearn.naive_bayes import GaussianNB, MultinomialNB , BernoulliNB
```

```python
gnb=GaussianNB()
gnb.fit(x_train , y_train)
```

```python
gnb.score(x_test, y_test)*100
```

```python
gnb.predict([[6.8, 123]]) # it's good
```

```python
gnb.score(x_train, y_train)*100
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=gnb)
plt.show()
```

```python
mnb=MultinomialNB()
```

```python
mnb.fit(x_test, y_test)
```

```python
mnb.score(x_test, y_test)*100
```

```python
mnb.score(x_train, y_train)*100
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=mnb)
plt.show()
```

```python
brn=BernoulliNB()
```

```python
brn.fit(x_test, y_test)
```

```python
brn.score(x_test, y_test)*100
```

```python
brn.score(x_train, y_train)*100
```

```python
plt.figure(figsize=(4,3))
plot_decision_regions(x.to_numpy(),y.to_numpy(),clf=brn)
plt.show()
```

```python
# best one is GaussianNB  because it works perfect on normal distribution
```
