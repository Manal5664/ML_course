```python
import pandas as pd
import numpy as np
import seaborn as sns
from sklearn import linear_model
```

```python
df= pd.read_csv("student_scores.csv")
df.head(5)
```

```python
sns.lmplot(x='Hours', y='Score', data=df )
```

```python
reg= linear_model.LinearRegression()
```

```python
reg.fit(df[['Hours']], df['Score'])
```

```python
reg.predict([[2.5]])
```

```python
reg.coef_
```

```python
reg.intercept_
```
