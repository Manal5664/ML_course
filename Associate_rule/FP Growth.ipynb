```python
market=[]
for i in range(0,df.shape[0]):
    customer=[]
    for j in df.columns:
        if type(df[j][i])==str:
            customer.append(df[j][i])
    market.append(customer)
```

```python
market
```

```python
l=[]
for i in market:
    for j in i:
        l.append(j)
```

```python
import collections
```

```python
p=collections.Counter(l)
```

```python
p.keys()
```

```python
p.values()
```

```python
d={'Item Name': p.keys(), 'Values':p.values() }
```

```python
pd.DataFrame(d).sort_values(by=['Values'], ascending=False)
```

```python
from mlxtend.preprocessing.transactionencoder import TransactionEncoder
```

```python
tr=TransactionEncoder()
tr.fit(market)
```

```python
tr.transform(market)
```

```python
df1=pd.DataFrame(tr.transform(market), columns=tr.columns_)
```

```python
df1
```

```python
from mlxtend.frequent_patterns import fpgrowth
```

```python
fp=fpgrowth(df1, min_support=0.07, use_colnames=True, max_len=3).sort_values(by='support')
```
