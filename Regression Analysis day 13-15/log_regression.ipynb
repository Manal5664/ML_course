```python
import pandas as pd
import  matplotlib.pyplot as plt
import seaborn as sns
# from sklearn.modelselection import 
```

```python
df= pd.read_csv('Social_Network_Ads.csv')
df.head(3)
```

```python
df.drop(columns=['User ID','Gender','EstimatedSalary'], inplace=True)
df
```

```python
plt.figure(figsize=(4,3))
sns.scatterplot(x='Age' , y='Purchased', data=df)
plt.show()
```

```python
x=df[['Age']]
y=df['Purchased']
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y , test_size=0.2, random_state=42)
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
lr=LogisticRegression()
lr.fit(x_train, y_train)
```

```python
lr.score(x_test, y_test)*100
```

```python
lr.predict([[15]])
```

```python
plt.figure(figsize=(7,5))
sns.scatterplot(x='Age' , y='Purchased', data=df)
sns.lineplot(x='Age' , y=lr.predict(x) , data=df, color='red')
plt.show()
```

```python
plt.figure(figsize=(6,6))
sns.scatterplot(x='Age' , y='Purchased', data=df)
sns.scatterplot(x='Age' , y=lr.predict(x) , data=df, color='red')
plt.show()
```
