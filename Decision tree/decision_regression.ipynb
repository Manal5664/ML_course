```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor , plot_tree
# from sklearn.linear_model import 
import os
```

```python
import kagglehub

# Download latest version
path = kagglehub.dataset_download("codebreaker619/salary-data-with-age-and-experience")

print("Path to dataset files:", path)
```

```python
file= os.listdir(path)
file
```

```python
df=pd.read_csv(os.path.join(path ,'Salary_Data.csv' ))
```

```python
df.head(3)
```

```python
df.isnull().sum()
```

```python
sns.pairplot(data=df)
plt.show()
```

```python
x=df[['YearsExperience', 'Age']]
y=df['Salary']
```

```python
x_train,x_test,y_train, y_test=train_test_split(x , y, test_size=0.2 , random_state=42)
```

```python
dtr=DecisionTreeRegressor()
dtr.fit(x_train, y_train)
```

```python
dtr.score(x_test, y_test)*100
```

```python
dtr.score(x_train, y_train)*100
```

```python
plot_tree(dtr)
plt.show()
```
