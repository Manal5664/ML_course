```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.model_selection import train_test_split
from mlxtend.plotting import plot_decision_regions
```

```python
df=pd.read_csv('Social_Network_Ads.csv')
df.head(3)
```

```python
df.drop(columns=df[['Gender','User ID']], inplace=True)
```

```python
df.head()
```

```python
df.isnull().sum()
```

```python
x=df.iloc[:,:-1]
x.head()
```

```python
y=df['Purchased']
y.head(3)
```

```python
df['Purchased'].unique()
```

```python
sns.scatterplot(data=df , x='Age', y='EstimatedSalary', hue='Purchased')
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
ss=StandardScaler()
ss.fit_transform(x)
ss
```

```python
x= pd.DataFrame(ss.fit_transform(x),columns=x.columns)
x
```

```python
x_train,x_test, y_train, y_test = train_test_split(x, y, test_size=0.2, random_state=42)
```

```python
from sklearn.tree import DecisionTreeClassifier
```

# Gini

# Pre pruning

```python
dt=DecisionTreeClassifier(max_depth=2)
```

```python
dt.fit(x_train, y_train)
```

```python
dt.score(x_test, y_test)*100
```

```python
dt.score(x_train, y_train)*100
# before pre pruning model is overfitted because its work well i training but not is testing
```

```python
dt.predict([[19,19000]])
```

```python
dt.predict([[35,20000]])
```

```python
plot_decision_regions(x.to_numpy() , y.to_numpy() , clf=dt)
plt.show()
```

```python
from sklearn.tree import plot_tree
```

```python
plt.figure(figsize=(10,20))
plot_tree(dt)
plt.show()
```

```python
x_train.shape
```

## Entropy

```python
dt1=DecisionTreeClassifier(criterion='entropy' , )
dt1.fit(x_train, y_train)
```

```python
dt1.score(x_test , y_test)*100
```

```python
plot_decision_regions(x.to_numpy() , y.to_numpy() , clf=dt1)
plt.show()
```

#  Post pruning

```python
for i in range(1, 20):
    dt2= DecisionTreeClassifier(max_depth=i)
    dt2.fit(x_train, y_train)
    print(dt2.score(x_train, y_train),dt2.score(x_test, y_test), i)
```
